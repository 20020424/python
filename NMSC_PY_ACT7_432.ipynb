```python
import random
from IPython.display import clear_output

#1.- Función que utilice una lista con los nombres de tus mascotas, o artistas favoritos, 
#(minimo 5, maximo 10) imprimir las cadenas y la cantidad de caracteres de cada cadena.  

def lista_nombres(lista_nombres):
    clear_output()
    print("1-. LISTA DE NOMBRES")
    for i,nombre in enumerate(lista_nombres):
        print(f"{[i]}{nombre}: {str(len(nombre))} caracteres")

nombres=["NANCY", "MARY", "SETH", "ALEX", "MARCELO", "SAMANTHA", "ALEXIS", "ABRAHAM"]
lista_nombres(nombres)

# 2-. Programa que utilice 2 funciones, 
#Función que genere y regrese una lista con 10 números aleatorios entre el 30 y 50 (no repetidos). 
#Función que reciba una lista , Imprimir la lista (Indice y Valor )

import random
def generar_lista_aleatoria():
    clear_output()
    lista = []
    while len(lista) < 10:
        numero_aleatorio = random.randint(30, 50)
        if numero_aleatorio not in lista:
            lista.append(numero_aleatorio)
    return lista

def imprimir_lista_con_indices(lista):
    clear_output()
    for indice, valor in enumerate(lista):
        print("Indice {}: {}".format(indice, valor))

lista_aleatoria = generar_lista_aleatoria()
imprimir_lista_con_indices(lista_aleatoria)

#3.- Escribe una función que reciba dos listas de números del mismo tamaño y calcule la suma de los elementos correspondientes de cada lista. 
#(regresar una nueva lista) . Luego, muestra una lista con los resultados de cada suma. Nota: si las listas no son del mismo tamaño mandar msge 
#y utilizar el tamaño de la lista mas pequeña).

def sumar_listas():
    clear_output()
    try:
        lista1 = [int(x) for x in input("Ingresa los numeros de la primera lista separados por espacios: ").split()]
        lista2 = [int(x) for x in input("Ingresa los numeros de la segunda lista separados por espacios: ").split()]
        minimo_tamano = min(len(lista1), len(lista2))
        lista1 = lista1[:minimo_tamano]
        lista2 = lista2[:minimo_tamano]
        lista_resultado = []
        for i in range(minimo_tamano):
            suma = lista1[i] + lista2[i]
            lista_resultado.append(suma)
        return lista_resultado
    except:
        print("Las listas no tienen el mismo tamaño o contienen elementos no numericos.")

lista_resultado = sumar_listas()
if lista_resultado:
    print(lista_resultado)

#4.- Escribe una función llamada eliminar_duplicados que reciba una lista como parámetro y elimine los elementos duplicados. 
#El resultado debe ser una nueva lista sin duplicados. Pide al usuario que ingrese una lista 
#y luego llama a la función eliminar_duplicados para mostrar la lista sin duplicados. (Solo Numeros enteros, validar)

def eliminar_duplicados(lista):
    clear_output()
    return list(set(lista))

while True:
    try:
        lista = input("Ingresa una lista de números enteros separados por espacios: ").split()
        lista = [int(num) for num in lista]
        break
    except ValueError:
        print("Error: ingresa solo números enteros separados por espacios.")

lista_sin_duplicados = eliminar_duplicados(lista)

print("Lista original:", lista)
print("Lista sin duplicados:", lista_sin_duplicados)

#5.- Escribe una función que calcule la media y la mediana de una lista de números enteros. 
#La media es la suma de todos los elementos y dividido entre la cantidad de elementos. 
#La mediana es el valor que queda en la mitad de la lista cuando se ordena de forma ascendente. Si la lista tiene un número par de elementos, 
#la mediana se calcula como el promedio de los dos valores centrales. Puedes utilizar el método sort() para ordenar la lista 
#y luego calcular la mediana según el tamaño de la lista.

def media_mediana():
    clear_output()
    lista = []
    
    while True:
        try:
            n = int(input("Ingrese un número entero a la lista (ingrese -1 para terminar): "))
            if n == -1:
                break
            lista.append(n)
        except ValueError:
            print("Debe ingresar un número entero.")
            
    media = sum(lista) / len(lista)

    lista.sort()
    if len(lista) % 2 == 0:
        mediana = (lista[len(lista)//2] + lista[len(lista)//2 - 1]) / 2
    else:
        mediana = lista[len(lista)//2]

    print("La lista ingresada es:", lista)
    print("La media es:", media)
    print("La mediana es:", mediana)

#6.- Escribe una función que calcule la media móvil de una lista de números enteros. 
#La media móvil es el promedio de un número determinado de elementos consecutivos de la lista. 
#Puedes utilizar unciclo y una variable de suma para calcular la media móvil de cada sublista y guardar los resultados en otra lista.

def media_movil():
    clear_output()
    try:
        lista = input("Ingresa una lista de números separados por comas: ").split(",")
        lista = [int(num) for num in lista]
        n = int(input("Ingresa el tamaño de la ventana de la media móvil: "))
        medias = []
        for i in range(len(lista) - n + 1):
            ventana = lista[i:i+n]
            media = sum(ventana) / n
            medias.append(media)
        return medias
    except ValueError:
        print("Error: Ingresa solo números enteros separados por comas.")


def menu():
    print("MENU:")
    print("1.-LISTA MASCOTAS")
    print("2.-FUNCION LISTA CON NUMEROS ALEATORIOS")
    print("3.-FUNCION LISTA INDICE Y VALOR")
    print("4.-SUMA LISTAS")
    print("5.-ELIMINAR DUPLICADOS DE UNA LISTA")
    print("6.-MEDIA Y MEDIANA DE UNA LISTA")
    print("7.-MEDIA MOVIL DE UNA LISTA")
    print("0.-SALIR")

#EJECUCION DE TODAS LAS FUNCIONES(MAIN)

while True:
    menu()
    op = int(input("Ingresa la funcion que deseas utilizar: "))
    if op == 1:
        
        lista_nombres(lista)
        input("Presiona ENTER para borrar la pantalla")
        clear_output()
        continue
    elif op == 2:
        generar_lista_aleatoria()
        input("Presiona ENTER para borrar la pantalla")
        clear_output()
        continue
    elif op == 3:
         imprimir_lista_con_indices(lista)
         input("Presiona ENTER para borrar la pantalla")
         clear_output()
         continue
    elif op == 4:
        sumar_listas()
        input("Presiona ENTER para borrar la pantalla")
        clear_output()
        continue
    elif op == 5:
        eliminar_duplicados(lista)
        input("Presiona ENTER para borrar la pantalla")
        clear_output()
        continue
    elif op == 6:
        media_mediana()
        input("Presiona ENTER para borrar la pantalla")
        clear_output()
        continue
    elif op == 7:
        media_movil()
        input("Presiona ENTER para borrar la pantalla")
        clear_output()
        continue
    elif op == 0:    
        break
```
